```python
#Ejercicio 1.1 - Una variable categòrica (UniqueCarrier)

carrier_counts = avion['UniqueCarrier'].value_counts()
plt.bar(carrier_counts.index,carrier_counts.values,color='C9')
plt.xlabel("Compañias aereas")
plt.ylabel("Número de vuelos")
plt.title("Vuelos por compañia aerea 2008 EEUU")
plt.savefig("C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T1/vuelosxcompañia.png")
plt.show()




```

```python
#Ejercicio 1.2 - Una variable numèrica (ArrDelay)
    #Seleccionamos la categoria deseada y la mostramos
retraso = avion[['ArrDelay']]
print(retraso)

    #Creamos el grafico de densidad con seaborn para ver donde esta la concentración respecto a horas de reptraso en el despegue
sns.kdeplot(data=retraso, x="ArrDelay")
plt.savefig("C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T1/densidad.png")
```

```python
#Ejercicio 1.3 - Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

    #Definimos un tamaño suficientemente grande para que se vea toda la información
plt.figure(figsize=(20,15))
    #Usamos seaborn para un boxplot de todas las compañias y retrasos en despeg
sns.boxplot(x="UniqueCarrier", y="ArrDelay", data=avion)
plt.title('Boxplot - retrasos despegue por compañia')
plt.savefig("C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T1/boxplot_uniquecarrier_arrdelay.png")
plt.show()
```

```python
#Ejercicio 1.4 - Dues variables numèriques (ArrDelay i DepDelay)
    #Seleccionamos las 2 variables
arr = avion['ArrDelay']
dep = avion['DepDelay']

    #Ordenamos por el index de 'arr' ('ArrDelay') y plasmamos
arr1 = arr.sort_index()
plt.subplot(3, 1, 1)
plt.title('Retrasos en el despegue')
plt.plot(arr.index,arr.values,"-")
plt.savefig("C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T1/retrasos_despegue.png")
plt.show()
    
    #Ordenamos por el index de 'dep' ('DepDelay') y plasmamos
dep1 = dep.sort_index()
plt.subplot(3, 1, 2)
plt.title('Retrasos en el aterrizaje')
plt.plot(dep.index,dep.values,"b*")
plt.savefig("C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T1/retrasos_aterrizaje.png")
plt.show()

    #Restamos arr - dep para obtener la diferencia entre despegue menos aterrizaje
dif = arr1-dep1
plt.subplot(3, 1, 3)
plt.title('Diferencia entre horas de despegue menos horas de aterrizaje')
plt.scatter(dif.index,dif.values)
plt.savefig("C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T1/retrasos_despegue-aterrizaje.png")
plt.show()

    #Vemos los datos
media = dif.mean()
maximo = dif.max()
minimo = dif.min()
print('El valor máximo es: ',maximo,', el valor mínimo es: ',minimo,' y el valor medio es: ',media)


#Si al restar salida - aterrizaje se obtiene :
    #un valor positivo quiere decir que hay mas horas de retraso en el aterrizaje que en la salida
    #un valor negativo quiere decir que hay menos horas de retraso en el aterrizaje que en la salida
    #un valor 0 quiere decir que hay las mismas horas de retraso en el aterrizaje que en la salida



```

```python
#Ejercicio 1.5 - Tres variables (ArrDelay, DepDelay i UniqueCarrier)
    #Filtramos datos por la compañia "WN" y seleccionamos solamente las columnas que queremos ver
WN = avion.loc[avion.UniqueCarrier=='WN',['UniqueCarrier','ArrDelay','DepDelay']]
TT = avion[['UniqueCarrier','ArrDelay','DepDelay']]



#compañias = WN.iloc[:,[0][0]]
#print(compañias)
#arr = WN.iloc[:,[1][0]]
#print(arr)
#dep = WN.iloc[:,[2][0]]
#print(dep)
 
#indice = np.arange(len(compañias))
TTsum = TT.groupby(['UniqueCarrier']).sum()
TTmean = TT.groupby(['UniqueCarrier']).mean()


TTsum.plot(kind = 'barh')
plt.xlabel("Horas de retraso (x7)")
plt.ylabel("Compañias aereas")
plt.title("Suma de horas de retrasos de despegues y aterrizajes")
plt.savefig("C:/Users/noel_/Dropbox/Python/data_science_python/Modulo 2/T1/suma_compañia.png")
plt.show()

TTmean.plot(kind = 'bar')
plt.xlabel("Compañias aereas")
plt.ylabel("Horas de retraso")
plt.title("Media de horas de retraso de despegues y aterrizajes")
plt.savefig("C:/Users/noel_/Dropbox/Python/data_science_python/Modulo 2/T1/media_compañia.png")
plt.show()



```

```python
#Ejercicio 1.6 - Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)
TT = avion[['UniqueCarrier','ArrDelay','DepDelay']]
TT2 = avion[['UniqueCarrier','AirTime']]


TTsum = TT.groupby(['UniqueCarrier']).mean()
TT2mean = TT2.groupby(['UniqueCarrier']).mean()

#plt.plot(TTsum.index,TTsum.values,color='C9')
TTsum[['ArrDelay','DepDelay']].plot(kind='bar')
TT2mean['AirTime'].plot(kind='line', color='black', marker='o')
plt.xlabel("Compañias aereas")
plt.ylabel("Horas de vuelo media/retraso")
plt.legend(['ArrDelay','DepDelay','ArrDelay'])
plt.title("H. de vuelo media/retraso por compañia aerea")
plt.savefig("C:/Users/noel_/Dropbox/Python/data_science_python/Modulo 2/T1/3variables.png")
plt.show()
```

```python
#Ejercicio 3 - Integra les visualitzacions gràfiques, en la tasca 2, del Mòdul 1.
mes = avion.iloc[:,[2,10]]
mes1 = mes.groupby(['Month']).count()
mes2 = mes1.sort_values(by=(['FlightNum']),ascending=False)
print(mes2)

mes2[['FlightNum']].plot(kind='barh')
plt.ylabel("Compañias aereas")
plt.xlabel("Número de vuelos")
plt.title("Número de vuelos por compañia")
plt.show()
```

```python
#Ejercicio 3 - Integra les visualitzacions gràfiques, en la tasca 2, del Mòdul 1.
#Funciona! Ranking en número de horas en retrasos por compañia
retraso = avion.iloc[:,[9,29]]
retraso1 = retraso.groupby('UniqueCarrier').sum().rank(ascending=False)

#añadimos una columna nueva con el total de horas de retraso y ordenamos por horas de retraso
retraso1.loc[:,'horas de retraso']= retraso.groupby('UniqueCarrier').sum()
retraso2 = retraso1.sort_values(by=(['LateAircraftDelay']),ascending=True)

print(retraso2)

retraso2[['horas de retraso']].plot(kind='bar')
plt.ylabel("Compañias aereas")
plt.xlabel("Número de vuelos")
plt.title("Número de vuelos por compañia")
plt.show()
```
